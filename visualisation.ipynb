4 Visualisation

```
import sys
import numpy as np
import matplotlib.pyplot as plt
```

4.1 Simple 2D plotting

```
x=np.arange(-2,2,0.01)
y=np.square(x)
plt.plot(x, y)
plt.show()
```

```

nx=np.size(x)
print(nx,x[1],x[nx-1])
```

```
y2=2*abs(np.sin(x))
#plt.style.use('classic')
plt.plot(x,y,'b',x,y2,'r:')
plt.xlabel('x')
plt.ylabel('y (blue) and y2 (red)')
plt.title('Demonstration of simple 2D-plotting')
plt.savefig('plot2D.png')
plt.show()
```

4.2 3D plotting

```
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
t=np.arange(0,20*np.pi,0.01)
x=np.cos(t)
y=np.sin(t)
z=np.power(t,3)
mpl.rcParams['legend.fontsize'] = 10
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x, y, z, label='parametric curve')
#ax.legend()
plt.xticks(np.arange(-1,1.1,0.5))
plt.yticks(np.arange(-1,1.1,0.5))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example of a 3D-plot')
plt.savefig('plot3Dlinje.png')
plt.show()

```

```
x=np.arange(-2,2,.2)
y=np.arange(-2,2,.2)
X, Y = np.meshgrid(x,y)
Z=np.exp(-np.square(X)-np.square(Y))
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,Y,Z)
plt.xticks(np.arange(-2,2,0.5))
plt.yticks(np.arange(-2,2,0.5))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example of a 3D-meshplot')
plt.savefig('plot3Dlinje.png')
plt.show()
```
